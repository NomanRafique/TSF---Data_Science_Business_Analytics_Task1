**importing Laibraires For Dtaa Visualizing**


```
import seaborn as sns
import matplotlib.pyplot as plt
```

```
x = data.Hours
y = data.Scores
```

```
sns.lineplot( x,y)
import warnings
# for ignoring warnings
warnings.filterwarnings("ignore")
```

```
plt.scatter(x,y, color  = '#88c999')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours STudent Studied')  
plt.ylabel('Percentage Student Scored')  
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.figure(figsize=(26,13))
plt.show()
```

```
x=data.iloc[:, :-1].values
y=data.iloc[:, 1].values
print("Shape of Number of Hours STudents Study",x.shape)

print("Shape of Percentage Student Scored",y.shape)
```

  **SPlitting**   

```

# Using in-built method of sci-kit learn of train_test_split()
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.4, random_state = 0)

print("Hours in train data ",x_train.shape)
print("Score in train data",y_train.shape)
print("Hours in Test data ",x_test.shape)
print("Score in Test data",y_test.shape)
 
```

**Model Training And Testing**

```
from sklearn.linear_model import LinearRegression
linear = LinearRegression()
linear.fit(x_train, y_train)
print("Model Has trainded Successfully!")
```

```
pred = linear.predict(x)
plt.scatter(x,y)
plt.plot(x,pred)
plt.show()
```

**Prediction the testing data & Comparing to Actual Data**

```
y_pred = linear.predict(x_test)
s1 = pd.Series(y_test)
s2 = pd.Series(y_pred)
pd.concat([s1, s2],axis=1, keys=['Actual','Prdicted'])
```

**For  Checking the accuracy/Score of Model**

```
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
```

```

print('Test Score')
print(linear.score(x_test, y_test))
print('Training Score')
print(linear.score(x_train, y_train))
```

```
from sklearn import metrics
mean_squ_error = mean_squared_error(y_test, y_pred )
mean_abs_error = mean_absolute_error(y_test, y_pred )
print("Mean Squared Error is:",mean_squ_error)
print("Mean Absolute Error is:",mean_abs_error)
```

```

print('Score of the student who studied for 9.25 hours is: ', linear.predict([[9.25]]))
```

**For Checking Student will pass or Fail**

```
number = int(input("Enter The Number of Hours for knowing student will be pass or fail ...Passing Percentage is 35%"))
percentage = linear.predict([[number]])
if percentage>35:
  print(f"Student is Pass Percentage is {percentage} ")
else :
  print(f"Student is Fail Percentage is {percentage} ")


```
